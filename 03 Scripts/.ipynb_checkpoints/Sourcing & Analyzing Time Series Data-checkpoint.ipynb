#### 05. Testing for stationarity

- Conduct a Dickey-Fuller test to check for stationarity

```python
# The adfuller() function will import from the model from statsmodels for the test; however, running it will only return 
# an array of numbers. This is why you need to also define a function that prints the correct output from that array.

from statsmodels.tsa.stattools import adfuller # Import the adfuller() function

def dickey_fuller(timeseries): # Define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the test using the function on the time series
dickey_fuller(data_sub['overall_score'])
```

- The test Statistic value is -0.11 which is larger than critical value of 5% (-3.10), which mean we canâ€™t reject null hypothesis. this result indicat that there is a unit root in the data and it is non-stationary.

#### 06. Stationarizing the Data

```python
#perform differencing
data_diff = data_sub - data_sub.shift(1) # The df.shift(1) function turns the observation to t-1, making the whole thing t - (t -1)
```

```python
data_diff.dropna(inplace = True) # Here, you remove the missing values that came about as a result of the differencing. 
# You need to remove these or you won't be able to run the Dickey-Fuller test.
```

```python
data_diff.head()
```

```python
data_diff.columns
```

```python
# Check out what the differencing did to the time-series curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
```

```python
#Repeat the dickey-fuller test
dickey_fuller(data_diff)
```

```python
# perform differencing second time
data_diff2 = data_diff - data_diff.shift(1)
```

```python
data_diff2.dropna(inplace = True)
```

```python
#Repeat the dickey-fuller test
dickey_fuller(data_diff2)
```

```python
#perform differencing third time
data_diff3 = data_diff2 - data_diff2.shift(1)
```

```python
data_diff3.dropna(inplace = True)
```

```python
#Repeat the dickey-fuller test
dickey_fuller(data_diff3)
```

```python
#perform differencing fourth time
data_diff4 = data_diff3 - data_diff3.shift(1)
```

```python
data_diff4.dropna(inplace = True)
```

```python
dickey_fuller(data_diff4)
```

```python
#perform differencing fifth time
data_diff5 = data_diff4 - data_diff4.shift(1)
data_diff5.dropna(inplace = True)
dickey_fuller(data_diff5)
```

- The test Statistic value is -5.35 after repeatedly performing differencing five times which is smaller than critical value of 5% and 10%, which mean we can reject null hypothesis now. this result indicat that there is no unit root in the data and it is stationary.

```python
# Check out what the differencing did to the time-series curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff5)
```
