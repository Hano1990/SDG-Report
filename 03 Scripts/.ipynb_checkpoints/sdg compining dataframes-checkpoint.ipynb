```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Importing SDG 2023 clean dataframe
sdg_2023=pd.read_csv(('../02 Data/Prepared Data/sdg_2023_clean.csv'),index_col=0)
sdg_2023.head()
```

```python
# Importing SDG 2000-2022 clean dataframe
sdg_2022=pd.read_csv(('../02 Data/Prepared Data/sdg_2000-2022_clean.csv'),index_col=0)
sdg_2022.head()
```

```python
# Merge sdg_2023 and sdg_2022 on "order_id" column
df_merged=pd.concat([sdg_2022,sdg_2023],ignore_index=True)
df_merged
```

```python
df_merged.isnull().sum()
```

```python
#change data type
df_merged['year']=df_merged['year'].astype('category')
```

```python
df_merged.describe()
```

```python
# Export df_merged data to csv
df_merged.to_csv('../02 Data/Prepared Data/sdg merged.csv')
```
