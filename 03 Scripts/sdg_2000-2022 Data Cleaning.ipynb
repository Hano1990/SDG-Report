```python
#change data type
sdg_2022['country_code']=sdg_2022['country_code'].astype('category')
sdg_2022['country']=sdg_2022['country'].astype('category')
sdg_2022['year']=sdg_2022['year'].astype('category')
```

```python
#Check for mixed data type
for col in sdg_2022.columns.tolist():
  weird = (sdg_2022[[col]].applymap(type) != sdg_2022[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (sdg_2022[weird]) > 0:
    print (col)
```
